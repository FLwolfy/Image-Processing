- Image Processing

```python
# Gray Scale
img = Image(256, 256, 3)
img.load("images/building2_color.raw")
gray_img = Image.gray_scale(img)
gray_img.save("output/building2_gray.raw")

show_images([img, gray_img], ["Original Image", "Gray Image"])
```

```python
# Watermark
watermark = Image(128, 128, 3)
watermark.load("images/dku_logo_color.raw")
watermarked_img = Image.water_mark(img, watermark, -10, 10, 220, 0.5)

watermark_gray = Image(128, 128, 1)
watermark_gray.load("images/dku_logo_gray.raw")
watermarked_gray_img = Image.water_mark(gray_img, watermark_gray, -10, 10, 220, 0.5)

show_images([watermarked_img, watermarked_gray_img], ["Watermarked Original Image", "Watermarked Gray Image"])
```
